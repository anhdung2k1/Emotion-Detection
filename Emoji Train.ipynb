```python
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import cv2
from keras.models import load_model
from keras.optimizers import RMSprop,SGD,Adam
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
```

```python
train_dir ='/content/drive/MyDrive/Đồ án ML/File datasets/Emoji/train'
val_dir = '/content/drive/MyDrive/Đồ án ML/File datasets/Emoji/test'
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(48,48),
        batch_size=64,
        color_mode="grayscale",
        class_mode='categorical')

validation_generator = val_datagen.flow_from_directory(
        val_dir,
        target_size=(48,48),
        batch_size=64,
        color_mode="grayscale",
        class_mode='categorical')
```

```python
checkpoint = ModelCheckpoint("/content/drive/MyDrive/Đồ án ML/File colab/Dũng/Emotion Recognition/model.h5", monitor='val_acc', verbose=1, save_best_only=True, mode='max')
```

```python
model = load_model('/content/drive/MyDrive/Đồ án ML/File colab/Dũng/Emotion Recognition/model.h5')
```

```python
model.load_weights('/content/drive/MyDrive/Đồ án ML/File colab/Dũng/Emotion Recognition/model_emotion.h5')
```

```python
early_stopping = EarlyStopping(monitor='val_loss',
                          min_delta=0,
                          patience=3,
                          verbose=1,
                          restore_best_weights=True
                          )

reduce_learningrate = ReduceLROnPlateau(monitor='val_loss',
                              factor=0.2,
                              patience=3,
                              verbose=1,
                              min_delta=0.0001)
callbacks_list = [early_stopping,checkpoint,reduce_learningrate]
```

```python
model.fit(
    train_generator,
    steps_per_epoch= train_generator.n // train_generator.batch_size,
    epochs = 50,
    validation_data = validation_generator,
    validation_steps = validation_generator.n // validation_generator.batch_size,
    callbacks = callbacks_list
)
```

```python
model.save_weights('/content/drive/MyDrive/Đồ án ML/File datasets/model_emotion.h5')
print("Saved Successful")
```
